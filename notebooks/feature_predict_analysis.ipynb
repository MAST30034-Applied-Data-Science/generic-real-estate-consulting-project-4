```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r"./data/curated/history_info.csv", parse_dates=['year'], dtype={'SA2': str}, index_col=0)
dataset = dataset[dataset["SA2"] == "202011022"].reset_index(drop=True)
```

```python
dataset
```

```python
cand = list(dataset.columns)
cand.remove('SA2')
cand.remove('year')
```

```python
for i in cand:
    dta = dataset[i]
    dta.index = dataset["year"]
    fig = plt.figure()
    dta.plot(figsize=(12,8))
```
